# Results on Iris task using normal propability distributions to sample weights

```python
from wann_genetic.tools import ExperimentSeries
from util import accuracy_swarmplot
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
data_dir = "../data/iris_normal"
series = ExperimentSeries.from_spec_file(data_dir)
series.discover_data_dir()
df = series.assemble_stats()
df = df.sort_values(by=['distribution/mean', 'distribution/sigma'])
```

## Best Mean Accuracies (per individual, mean over sampled weights)
### best mean per interval

```python
df['mean,sigma'] = df.agg('{0[distribution/mean]},{0[distribution/sigma]}'.format, axis=1)
accuracy_swarmplot(data=df, x="mean,sigma", xlabel='$\mu,\sigma$', old_metrics=True)
```

```python
accuracy_swarmplot(data=df, x="distribution/mean", xlabel='$\mu$', old_metrics=True)
```

```python
accuracy_swarmplot(data=df, x="distribution/sigma", xlabel='$\sigma$', old_metrics=True)
```

The worst results were produced at high $\mu$ and low $\sigma$.

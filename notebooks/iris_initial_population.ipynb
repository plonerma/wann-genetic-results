# Influence of the initial population on the result of the experiment

The initially generated population is certainly important for the course of the experiment. Since it is created randomly we need to asess how much of an influence it will have on the end result.

In this experiment, three initial populations were created and then evolved using different post initialization seeds.

```python
from wann_genetic.tools import ExperimentSeries
import os
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
data_dir = "../data/iris_init_pop/trial 1"
series = ExperimentSeries.from_spec_file(data_dir)
series.discover_data_dir()
df1 = series.assemble_stats()
series.affected_keys()
```

```python
sns.swarmplot(x="post_init_seed", y="mean accuracy", data=df1)
plt.ylim([0,1])
plt.show()
```

```python
data_dir = "../data/iris_init_pop/trial 2"
series = ExperimentSeries.from_spec_file(data_dir)
series.discover_data_dir()
df2 = series.assemble_stats()
series.affected_keys()
```

```python
sns.swarmplot(x="population_seed", y="mean accuracy", data=df2)
plt.ylim([0,1])
plt.show()
```

```python
sns.swarmplot(x="initial_edge_prob", y="mean accuracy", data=df2)
plt.ylim([0,1])
plt.show()
```

```python
df = pd.concat([df1, df2])
```

```python
sns.swarmplot(x="initial_edge_prob", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.show()
```

```python
df['initial_edge_prob'] = df['initial_edge_prob'].map(lambda v: 'no edges' if np.isnan(v) else v)
sns.swarmplot(x="initial_edge_prob", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.show()
```

# Results on Iris task using uniform propability distributions to sample weights

```python
from wann_genetic.tools import ExperimentSeries, mean_comparison
from util import accuracy_swarmplot
import numpy as np
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
data_dir = "../data/iris_uniform"
series = ExperimentSeries.from_spec_file(data_dir)
series.discover_data_dir()
df = series.assemble_stats()
```

```python
df = df.sort_values(by=['distribution/lower_bound', 'distribution/upper_bound'])
df['lower_bound'] = df['distribution/lower_bound']
df['upper_bound'] = df['distribution/upper_bound']
```

## Best Mean Accuracies (per individual, mean over sampled weights)
### best mean per interval

```python
df['interval'] = df.agg('{0[lower_bound]},{0[upper_bound]}'.format, axis=1)
accuracy_swarmplot(data=df, x="interval", xlabel="Interval", old_metrics=True)
```

### best mean per lower bound type (sign of lower bound)

```python
df['lower_bound_type'] = np.sign(df['lower_bound'])
accuracy_swarmplot(data=df, x="lower_bound_type", xlabel="Sign of lower bound", old_metrics=True)
```

```python
mean_comparison(df, 'lower_bound_type', (-1,0,1))
```

### best mean per interval size

```python
df['interval_size'] = df['upper_bound'] - df['lower_bound']
accuracy_swarmplot(data=df, x="interval_size", xlabel="Interval size", old_metrics=True)
```

```python
mean_comparison(df, 'interval_size', (2,4,8,16))    
```

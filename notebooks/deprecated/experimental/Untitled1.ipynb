```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
linspace = np.linspace(0,1,100)
x, y = np.meshgrid(linspace, linspace)
```

```python
plt.imshow(np.fmin(x, y), origin='lower', vmin=0, vmax=1)
```

```python
plt.imshow(x*y, origin='lower', vmin=0, vmax=1)
```

```python
plt.imshow(x+y-2*x*y, origin='lower', vmin=0, vmax=1)
```

```python
plt.imshow((x-y)**2, origin='lower', vmin=0, vmax=1)
```

```python
plt.imshow(np.abs(x-y), origin='lower', vmin=0, vmax=1)
```

```python
plt.imshow(((x-y)**2 + x+y - 2*x*y)/2, origin='lower', vmin=0, vmax=1)
```

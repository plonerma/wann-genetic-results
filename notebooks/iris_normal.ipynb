# Results on Iris task using normal propability distributions to sample weights

```python
from rewann.tools import load_experiment_series
import os
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
df = load_experiment_series("../data/iris_normal")
df = df.sort_values(by=['distribution/mean', 'distribution/sigma'])
```

## Best Mean Accuracies (per individual, mean over sampled weights)
### best mean per interval

```python
df['mean,sigma'] = df.agg('{0[distribution/mean]},{0[distribution/sigma]}'.format, axis=1)
sns.swarmplot(x="mean,sigma", y="mean accuracy", data=df)
plt.xlabel('$\mu,\sigma$')
plt.ylim([0,1])
plt.show()
```

```python
sns.swarmplot(x="distribution/mean", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.xlabel('$\mu$')
plt.show()
```

```python
sns.swarmplot(x="distribution/sigma", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.xlabel('$\sigma$')
plt.show()
```

The worst results were produced at high $\mu$ and low $\sigma$.

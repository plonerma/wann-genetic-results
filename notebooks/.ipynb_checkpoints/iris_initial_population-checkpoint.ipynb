# Influence of the initial population on the result of the experiment

The initially generated population is certainly important for the course of the experiment. Since it is created randomly we need to asess how much of an influence it will have on the end result.

In this experiment, three initial populations were created and then evolved using different post initialization seeds.

```python
from rewann.tools.compare_experiments import assemble_dataframe
import os
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
base_path = "../data/iris_init_pop/trial 1"
spec_path = os.path.join(base_path, 'spec.toml')
data_path = os.path.join(base_path, 'data')
df = assemble_dataframe(spec_path, data_path,
         map_from_params={
            'initial population seed': ['sampling', 'seed'],
            'post initial pop seed': ['sampling', 'post_init_seed'],
         })
df = df.sort_values(by=['seed', 'post_init_seed'])
```

```python
sns.swarmplot(x="initial population seed", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.show()
```

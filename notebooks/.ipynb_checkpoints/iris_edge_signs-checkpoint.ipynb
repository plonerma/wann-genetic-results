```python
from rewann.tools import load_experiment_series, mean_comparison
import os
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import ttest_ind
import pandas as pd
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
df = load_experiment_series("../data/iris_edge_signs")
```

```python
sns.swarmplot(x="edge_signs_enabled", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.show()
```

```python
df['exp'] = df.agg('{0[distribution/_name]}, {0[seed]}'.format, axis=1)
df.index = df['exp']
a = df[df['edge_signs_enabled']]['mean accuracy']
b = df[~df['edge_signs_enabled']]['mean accuracy']
pd.DataFrame({'enabled': a, 'not enabled': b, 'diff': a-b})
```

```python
print(f'mean with edges signs: {a.mean()}, without: {b.mean()}')
from scipy.stats import ttest_ind
t, p = ttest_ind(a, b, equal_var=False)
print (t, p)
```

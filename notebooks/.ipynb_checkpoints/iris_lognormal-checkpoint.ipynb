```python
from rewann.tools.compare_experiments import load_experiment_series, mean_comparison
import os
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import ttest_ind
```

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

# Results on Iris task using lognormal propability distributions to sample weights
## Trial 1

```python
df = load_experiment_series(
    path="../data/iris_lognormal/trial 1",
    params_map=dict(
        mean=['sampling', 'mean'],
        sigma=['sampling', 'sigma'],
    )
)
df = df.sort_values(by=['mean', 'sigma'])
```

```python
df['mean,sigma'] = df.agg('{0[mean]},{0[sigma]}'.format, axis=1)
sns.swarmplot(x="mean,sigma", y="mean accuracy", data=df)
plt.xlabel('$\mu,\sigma$')
plt.ylim([0,1])
plt.show()
```

```python
sns.swarmplot(x="mean", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.xlabel('$\mu$')
plt.show()
```

```python
sns.swarmplot(x="sigma", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.xlabel('$\sigma$')
plt.show()
```

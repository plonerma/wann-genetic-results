# Results on Iris task using normal propability distributions to sample weights

```python
from rewann.tools.compare_experiments import assemble_dataframe
import os
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path = "../data/iris_normal"
spec_path = os.path.join(path, 'iris_normal_spec.toml')
df = assemble_dataframe(spec_path, path,
         map_from_params=dict(
            mean=['sampling', 'mean'],
            sigma=['sampling', 'sigma'],
         ))
df = df.sort_values(by=['mean', 'sigma'])
```

## Best Mean Accuracies (per individual, mean over sampled weights)
### best mean per interval

```python
df['mean,sigma'] = df.agg('{0[mean]},{0[sigma]}'.format, axis=1)
sns.swarmplot(x="mean,sigma", y="mean accuracy", data=df)
plt.xlabel('$\mu,\sigma$')
plt.ylim([0,1])
plt.show()
```

```python
sns.swarmplot(x="mean", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.xlabel('$\mu$')
plt.show()
```

```python
sns.swarmplot(x="sigma", y="mean accuracy", data=df)
plt.ylim([0,1])
plt.xlabel('$\sigma$')
plt.show()
```

The worst results were produced at high $\mu$ and low $\sigma$.
